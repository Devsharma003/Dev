```python
country_name = []
country_code = []
year = []
unemployment_rate = []
column_names = data.columns[2:]

for index, row in data.iterrows():
    for temp_year in column_names:
        country_name.append(row['Country Name'])
        country_code.append(row['Country Code'])
        year.append(temp_year)
        unemployment_rate.append(row[temp_year])

df2 = pd.DataFrame(data={'Year': year,'Country Name': country_name,'Country Code': country_code,'Unemployement Rate': unemployment_rate})
df2.head()
```

```python
fig = px.choropleth(df2, locations="Country Code",
                    color="Unemployement Rate",
                    hover_name="Country Name",
                    animation_frame="Year",
                    range_color=(0, 40),
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title='Interactive Map + Time Slider')
#fig.update_layout(transition = {'duration': 10})
fig.show()
```

This map visualizes the unemployment of all countries throughout time

It makes it easier to discover global or local crisis in time

Greenland and Antarctic are not in the dataset

```python
fig = px.line(df2, x="Year", y="Unemployement Rate", color='Country Name')
fig.show()
```

```python
country_names = df2['Country Name'].unique()
df2['gap_improvement'] = 0
start_years = ['1991','1992','1993','1994','1995']
end_years = ['2017','2018','2019','2020','2021']
high_improvement_countrys = df2.groupby('Country Name').mean('gap_improvement').sort_values('gap_improvement').index[:8]
low_improvement_countrys = df2.groupby('Country Name').mean('gap_improvement').sort_values('gap_improvement').index[-8:]
```

```python
fig = px.line(df2[df2['Country Name'].isin(low_improvement_countrys) ], x="Year", y="Unemployement Rate", color='Country Name', title='Best Country Developement')
fig.show()
```

```python
fig = px.line(df2[df2['Country Name'].isin(high_improvement_countrys) ], x="Year", y="Unemployement Rate", color='Country Name', title='Worst Country Developement')
fig.show()
```

```python
country_names = df2['Country Name'].unique()
df2['gap_minmax'] = 0
low_gap_countrys = df2.groupby('Country Name').mean('gap_minmax').sort_values('gap_minmax').index[:8]
high_gap_countrys = df2.groupby('Country Name').mean('gap_minmax').sort_values('gap_minmax').index[-8:]
```

```python
fig = px.line(df2[df2['Country Name'].isin(high_gap_countrys) ], x="Year", y="Unemployement Rate", color='Country Name', title='Highest Volatile Countries')
fig.show()
```

```python
fig = px.line(df2[df2['Country Name'].isin(low_gap_countrys) ], x="Year", y="Unemployement Rate", color='Country Name', title='Lowest Volatile Countries')
fig.show()
```
