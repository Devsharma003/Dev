```python
#creating pivot table
book_pivot=pd.pivot_table(data=book,values='Book.Rating',index='User.ID',columns='Book.Title')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot.head()
```

```python
book1 = 1 - pairwise_distances(book_pivot,metric='cosine')
book1
```

```python
book_data=pd.DataFrame(book1)
book_data.index= book_pivot.index
book_data.columns = book_pivot.index
book_data
```

```python
book_data.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(book1,0)
book_data.iloc[0:5,0:5]
```
