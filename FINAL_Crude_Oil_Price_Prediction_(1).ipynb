## Stationarity check using ADF

```python
from statsmodels.tsa.stattools import adfuller
print ('Results of Dickey-Fuller Test:')
dftest = adfuller(indexedData, autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```

#### Here, p-value is 0.616923 > 0.05. Hence, the data is not stationary.

```python
def MAPE(orig,pred):
    mape = np.mean(np.abs((orig-pred)/orig))*100
    return mape

def RMSE(orig, pred):
    rmse=np.sqrt(np.mean((np.array(orig)-np.array(pred))**2))
    return rmse
```

### Exponential Model

### Holt Model

```python
from statsmodels.tsa.api import Holt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
import warnings
warnings.filterwarnings("ignore")
```

```python
train_data = train["price_diff"]
test_data = test["price_diff"]

smoothing_levels = np.linspace(0.1, 1.0, num=10)
smoothing_slopes = np.linspace(0.1, 1.0, num=10)

best_mape = float('inf')
best_params = None

for alpha in smoothing_levels:
    for beta in smoothing_slopes:
        model = Holt(train_data).fit(smoothing_level=alpha, smoothing_slope=beta)
        forecast = model.predict(start=test_data.index[0], end=test_data.index[-1])
        mape = MAPE(forecast, test_data)

        if mape < best_mape:
            best_mape = mape
            best_params = (alpha, beta)

print("Best smoothing level (alpha):", best_params[0])
print("Best smoothing slope (beta):", best_params[1])
```

```python
hw_model = Holt(train["price_diff"]).fit(smoothing_level=0.2, smoothing_slope=0.8) #smoothing slope for trend
holt_pred = hw_model.predict(start = test.index[0],end = test.index[-1])
holt_rmse = RMSE(test['price_diff'],holt_pred)
holt_mape = MAPE(test['price_diff'],holt_pred)
```

```python
holt_rmse, holt_mape
```

### Holts winter exponential smoothing with multiplicative seasonality and additive trend

```python

train_log = np.log(train['price_diff'] + 2)  # Adding 2 to make the data strictly positive
test_log = np.log(test['price_diff'] + 2)

hwe_model_mul_add = ExponentialSmoothing(train_log, seasonal='mul', trend='add', seasonal_periods=12).fit()

pred_log_mul_add = hwe_model_mul_add.predict(start=test.index[0], end=test.index[-1])

holt_mul_add_pred = np.exp(pred_log_mul_add) - 2  # Subtracting 1 to revert the added constant

holt_mul_add_rmse = RMSE(test['price_diff'],holt_mul_add_pred)
holt_mul_add_mape = MAPE(test['price_diff'],holt_mul_add_pred)
```
